#### Import Libraries and Data

```python
#Import the libraries
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Load the dataset Flights with error handling in place
try:
    df = sns.load_dataset('flights')
except Exception as e:
    print(f"An error occured: {e}")
```

#### Data Exploration and Cleaning

```python
#Display the first few rows
df.head()
```

```python
#Display the column statistics
df.describe()
```

```python
#Display column types and any null values
df.info()
```

```python
#Drop any null values
df = df.dropna()
```

```python
#Compute the average passengers per month
df.groupby(by=['month'],observed=False)['passengers'].mean()
```

#### Visualization

```python
#Generate a histogram plot of the passengers
plt.title('Histogram of Number of Passengers')
plt.hist(x=df['passengers'])
plt.show()
```

```python
#Generate a line plot
plt.title('Number of Passengers per year')
sns.lineplot(data=df,x='year',y='passengers')
plt.show()
```

```python
#Compute the average per month
df_average_passengers = df.groupby(by=df['month'], observed=False)['passengers'].mean()

#Generate a bar plot on Month and average number of Passengers
plt.title('Average passengers per month')
sns.barplot(data=df_average_passengers)
plt.show()
```

```python
#generate a scatter plot
plt.title('Scatter plot of passengers by year')
sns.scatterplot(data=df, x = 'year', y = 'passengers')
plt.show()
```
